<h1>Introduction to Matplotlib</h1>
<li>Matplotlib is the most popular plotting library for Python</li>
<li>It gives you control over every aspect of a figure</li>
<li>It was designed to have a similar feel to MatLab's graphical plotting</li>
<li>You'll need to install it with pip or conda at your command line or terminal with:</li>

<code>conda install matplotlib</code>
<code>pip install matplotlib</code>

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y)
#plt.show()
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

#for current_ax in axes:
#    current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

<h2>Figure Size and DPI</h2>

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x**2)
ax.plot(x**3)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax.legend(loc=0)
```

<h2>Plot Appearance</h2>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', linewidth=20, alpha=0.5, linestyle='--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', linewidth=3, alpha=0.6, linestyle='--', marker='o')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=0.5, alpha=0.6, ls='-', marker='1', markersize=10)
```
